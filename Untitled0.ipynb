```python
# Mount Google Drive
from google.colab import drive
drive.mount('/content/drive')

# Import necessary libraries
import os
import zipfile
import numpy as np
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.efficientnet import EfficientNetB0, preprocess_input
from tensorflow.keras.models import Model

# Initialize EfficientNetB0 model for feature extraction
base_model = EfficientNetB0(weights='imagenet')
model = Model(inputs=base_model.input, outputs=base_model.get_layer('avg_pool').output)

def extract_cnn_features(model, image_path):
    img = image.load_img(image_path, target_size=(224, 224))
    img_array = image.img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0)
    img_array = preprocess_input(img_array)
    features = model.predict(img_array)
    return features.flatten()

# Paths to the zip files in Google Drive
zip_files = {
    "UERD": '/content/drive/My Drive/DATASETS/UERD.zip',
    "Test": '/content/drive/My Drive/DATASETS/Test.zip',
    "JUNIWARD": '/content/drive/My Drive/DATASETS/JUNIWARD.zip',
    "JMiPOD": '/content/drive/My Drive/DATASETS/JMiPOD.zip',
    "Cover": '/content/drive/My Drive/DATASETS/Cover.zip'
}

# Unzip files and process images
features_list = []
labels_list = []

for category, zip_path in zip_files.items():
    # Unzip the folder
    unzip_folder = f'/content/{category}'
    with zipfile.ZipFile(zip_path, 'r') as zip_ref:
        zip_ref.extractall(unzip_folder)

    # Process each image in the unzipped folder
    for img_file in os.listdir(unzip_folder):
        img_path = os.path.join(unzip_folder, img_file)
        features = extract_cnn_features(model, img_path)
        features_list.append(features)
        labels_list.append(category)

# Convert the lists to NumPy arrays
features_array = np.array(features_list)
labels_array = np.array(labels_list)

# You can now use features_array and labels_array for your machine learning tasks

# ... [previous code remains the same]

# Modified loop for batch processing and progress monitoring
batch_size = 100  # Process 100 images at a time
total_images = sum([len(os.listdir(f'/content/{category}')) for category in zip_files])
processed_images = 0

for category, zip_path in zip_files.items():
    unzip_folder = f'/content/{category}'
    image_files = os.listdir(unzip_folder)

    for i in range(0, len(image_files), batch_size):
        batch_files = image_files[i:i+batch_size]

        for img_file in batch_files:
            img_path = os.path.join(unzip_folder, img_file)
            features = extract_cnn_features(model, img_path)
            features_list.append(features)
            labels_list.append(category)

        processed_images += len(batch_files)
        print(f"Processed {processed_images}/{total_images} images")

# Convert the lists to NumPy arrays
features_array = np.array(features_list)
labels_array = np.array(labels_list)
```

```python
from sklearn.model_selection import train_test_split

# Assuming you have loaded your labels_array from the .npy file already
# Update the label_mapping to include all your labels
label_mapping = {
    'Cover': 0,
    'JMiPOD': 1,
    'JUNIWARD': 1,
    'UERD': 1,
    'Test': 1  # Update this if 'Test' should not be labeled as 1
}

# Map the labels to binary
y_binary = np.array([label_mapping[label] for label in labels_array])

# Split the data
X_train, X_val, y_train, y_val = train_test_split(features_array, y_binary, test_size=0.2, random_state=42)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout

# Building the neural network model
model = Sequential()
model.add(Dense(512, activation='relu', input_shape=(X_train.shape[1],)))
model.add(Dropout(0.5))
model.add(Dense(1, activation='sigmoid'))

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train the model
model.fit(X_train, y_train, epochs=10, batch_size=32, validation_data=(X_val, y_val))
```

```python
from sklearn.model_selection import KFold
import numpy as np
import tensorflow as tf

def create_model():
    model = tf.keras.Sequential([
        tf.keras.layers.Dense(512, activation='relu', input_shape=(X_train.shape[1],)),
        tf.keras.layers.Dropout(0.5),
        tf.keras.layers.Dense(1, activation='sigmoid')
    ])
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model

# Prepare K-Fold cross-validation
kfold = KFold(n_splits=5, shuffle=True, random_state=42)

acc_scores = []
loss_scores = []

for train, test in kfold.split(features_array, y_binary):
    model = create_model()
    model.fit(features_array[train], y_binary[train], epochs=10, batch_size=32, verbose=0)
    loss, acc = model.evaluate(features_array[test], y_binary[test], verbose=0)
    acc_scores.append(acc)
    loss_scores.append(loss)

print(f"Average accuracy: {np.mean(acc_scores)}")
print(f"Average loss: {np.mean(loss_scores)}")
```
